```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
ronalod_img = cv2.imread("images\Ronaldo.jpg")
eyes_img = cv2.imread("images\\brown-eyes.jpg")
```

```python
img = eyes_img.copy()
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
# gauss = cv2.GaussianBlur(gray,(15,15),5)
circles = cv2.HoughCircles(gray,cv2.HOUGH_GRADIENT,dp=1,minDist=90,minRadius=1,maxRadius=70,param1=180,param2=40)
circles = np.uint16(np.around(circles))
print(circles)
for circle in circles[0]:
    centre = (circle[0],circle[1])
    radius = circle[2]
    cv2.circle(img,centre,radius,(0,255,0),2)

plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()


```

```python
img = ronalod_img.copy()
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
# gauss = cv2.GaussianBlur(gray,(15,15),5)
circles = cv2.HoughCircles(gray,cv2.HOUGH_GRADIENT,dp=1,minDist=90,minRadius=100,maxRadius=145,param1=180,param2=40)
circles = np.uint16(np.around(circles))
print(circles)
for circle in circles[0]:
    centre = (circle[0],circle[1])
    radius = circle[2]
    cv2.circle(img,centre,radius,(0,0,255),10)

plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
